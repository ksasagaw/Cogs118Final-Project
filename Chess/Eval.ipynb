```python
import pandas as pd
import subprocess
import numpy as np
import time
```

```python
df = pd.read_csv('games.csv')
df = df[df['turns'] > 20]
df = df[df['turns'] < 50]
df = df[df['rated'] == True]
df = df[df['victory_status'] == 'mate']
df = df[['moves', 'turns']]
move_list = df['moves']
temp_list = []
for x in range(10, df['turns'].max()):
    temp_list.append('move' + str(x))
df = df.reindex(columns = df.columns.tolist() + temp_list)
print(df)

for idx, moves in move_list.iteritems():
    count = 10
    for y in range(10, len(moves.split())):
        name = 'move' + str(count)
        mveList = ''
        for x in range(0,count):
            if(x == 0):
                mveList += moves.split()[x]
            else:
                mveList += ' ' + moves.split()[x]
        df.loc[idx, name] = mveList
        
        count += 1
```

```python
df.to_csv('movelist.csv')
df
```

```python
mv_df = pd.read_csv('movelist.csv',index_col=0)
temp_list = []
for x in range(10, int(df['turns'].max())):
    temp_list.append('uniq move' + str(x))
mv_df = mv_df.reindex(columns = mv_df.columns.tolist() + temp_list)
for x in range(10, int(mv_df['turns'].max())):
    name = 'move' + str(x)
    unq_name = 'uniq move' + str(x)
    mv_df[unq_name] = pd.Series(mv_df[name].unique()).dropna()
    
mv_df = mv_df[mv_df.columns[41:]]
mv_df = mv_df.dropna(how = 'all')
```

```python
np.sum(mv_df.count())
```

```python
for x in mv_df:
    for idx, moves in mv_df[x].iteritems():
        if(str(moves) == 'nan'):
            print('pass')
            break
        temp_moves = str(moves) + " \*"
        subprocess.run("echo " + temp_moves + " > ~/Chess/temp_pgn.txt", shell = True, cwd  = "/home/ksasagaw/Chess", capture_output=True)
        pgn = subprocess.run("./pgn-extract -Wuci ~/Chess/temp_pgn.txt", shell = True, cwd  = "/home/ksasagaw/Chess/pgn-extract", capture_output=True)
        pgn = str(pgn.stdout).split('\\n')[8]
        pgn = pgn.replace('*','')
        mv_df.at[idx,x] = pgn
    print(x)
```

```python
mv_df.to_csv('uci_data.csv')
```

```python
#popen referenced https://eli.thegreenplace.net/2017/interacting-with-a-long-running-child-process-in-python/
uci_df = pd.read_csv('uci_data.csv', index_col = 0)
a = subprocess.Popen("./stockfish", cwd = "/home/ksasagaw/Chess/Stockfish/src", shell = True, stdin=subprocess.PIPE, stdout=subprocess.PIPE)
a.stdin.write('uci\n'.encode())
a.stdin.flush()
a.stdin.write('isready\n'.encode())
a.stdin.flush()
a.stdin.write('setoption name UCI_AnalyseMode value true\n'.encode())
a.stdin.flush()
print('set up')
for x in range(1,30):
    a.stdout.readline() #read every set up line

for x in uci_df:
    for idx, moves in uci_df[x].iteritems():
        print(moves)
        if(str(moves) == 'nan'):
            break
        a.stdin.write(('position startpos moves ' + str(moves) + '\n').encode())
        a.stdin.flush()
        a.stdin.write('go depth 20\n'.encode())
        a.stdin.flush()
        
        eval_line, best_line = '',''
        while('bestmove' not in best_line):
            time.sleep(.12)
            eval_line = best_line
            best_line = str(a.stdout.readline())
        score = eval_line.split(' ')[9]
        best_move = best_line.split(' ')[1]
        uci_df.at[idx,(x + '_score')] = score
        uci_df.at[idx,(x + '_bestmove')] = best_move
        print(score, best_move)
    print(x)
```
